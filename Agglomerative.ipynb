#### Agglomerative Clustering

```python
from  warnings import filterwarnings
filterwarnings('ignore')
```

#### Read the mall customers dataset

```python
import pandas as pd
df = pd.read_csv('Mall_Customers.csv')
df
```

```python
df.info()
```

```python
df.columns = ['CustomerID','Gender','Age','Income','Spending']
```

```python
df
```

#### Get X value

```python
X = df[['Age','Income','Spending']]
X
```

#### Perform Preprocessing on X

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
```

```python
num_pipe = Pipeline(steps=[('imputer',SimpleImputer(strategy='median')),
                           ('scaler',StandardScaler())])
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre[0:5]
```

```python
cols = num_pipe.get_feature_names_out()
cols
```

```python
X_pre = pd.DataFrame(X_pre,columns=cols)
X_pre.head()
```

#### Finding optimal number of cluster with Silhoutee score

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import AgglomerativeClustering
```

```python
n = list(range(2,16))
n
```

```python
scores = []
for i in n:
    agc = AgglomerativeClustering(n_clusters=i)
    agc.fit(X_pre)
    lb = agc.labels_
    sh = silhouette_score(X_pre, lb)
    scores.append(sh)
```

```python
scores
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
plt.plot(n,scores)
plt.xlabel('n_clusters')
plt.ylabel('silhouette_score')
plt.title('Optimal number of clusters')
plt.show()
```

```python
model = AgglomerativeClustering(n_clusters=6)
model.fit(X_pre)
```

```python
lb = model.labels_
lb
```

#### Giving labels for X data

```python
X['Label'] = lb
```

```python
X
```

#### Show different clusters in the table

```python
for i in range (6):
    print(f'Cluster Number : {i}')
    display(X[X['Label']==i])
    print('n\=================================\n')
```

#### Centroids

```python
C = X.groupby(by='Label').mean()
C
```

#### Checking the dendogram

```python
from scipy.spatial import distance_matrix
dm = distance_matrix(X_pre, X_pre)
dm
```

```python
dm.shape
```

```python
from scipy.cluster.hierarchy import linkage
linkages = linkage(dm)
```

```python
from scipy.cluster.hierarchy import dendrogram
Q = dendrogram(linkages)
```
